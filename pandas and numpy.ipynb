```python
df = df.sample(frac=1)
#df의 파일을 1비율로 랜덤하게 샘플링
```

```python
df = df.sort_values(by='Seed', ascending=True)
#오름차순 정렬 
df
```

```python
df['draw']=0
#'columm 명' = value  // 열을 하나 생성 
df
```

```python
arr = df.values 
# array 변경 

print(arr)
```

```python
# 조추첨
# 1. 같은리그일 경우 다시 돌림 

# 0~7행 : 2위팀
# 8~15행 : 1위팀 

re = 0
while re ==0:
    for i in range(8):
        #마지막 열 draw를 1로 변경 
        arr[i, -1] =1
        for j in np.arrange(i+8, 16):
            #조가 같지 않아야 한다. , 리그가 같지 않아야한다, 안뽑았을 때 추첨한다. 
            if(arr[j,1] != arr[i,1] and (arr[j,3] != arr[i,3]) and (arr[j,-1]==0)):
                arr[j,-1]=1
                #swap
                #0행 vs 8행, 1행 vs 9행, ... , 7행 vs 15행 
                arr[[i+8,j]] = arr[[j, i+8]]
                break 
        # 맞는지 검산 
        # 마지막에 뽑힌 팀들이 같은 조에 있었는지 or 같은 리그에 있는지 확인 
        if arr[-1,1]==arr[7,1] or (arr[-1,3]==arr[7,3]):
            re=0
        else:
            re=1 
```

```python
for i in range(8):
    print(f'{arr[i,0]}({arr[i,1],arr[1,3]}) vs {arr[i+8,0]}({arr[i+8,1],arr[i+8,3]})')
```
